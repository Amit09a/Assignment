1. What is hypothesis testing in statistics?
- Hypothesis testing in statistics is a formal method used to make decisions or draw conclusions about a population based on sample data. It begins with formulating a null hypothesis, which represents the default assumption of no effect or no difference, and an alternative hypothesis, which represents the claim being tested. Using sample data, a suitable test statistic is calculated and compared against a significance level (commonly 0.05) through a p-value. The p-value indicates the probability of observing the collected data, or something more extreme, assuming the null hypothesis is true. If this probability is sufficiently small, the null hypothesis is rejected; otherwise, it is not rejected. Hypothesis testing provides a systematic and objective framework for evaluating uncertainty and is widely applied in scientific research, data analysis, and decision-making processes.

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?

- The null hypothesis (H‚ÇÄ) is a statistical statement that assumes there is no effect, no difference, or no relationship in the population and serves as the default or starting assumption in hypothesis testing. In contrast, the alternative hypothesis (H‚ÇÅ or H‚Çê) represents the opposing claim that there is an effect, a difference, or a relationship and reflects what the researcher aims to demonstrate. While the null hypothesis is tested directly using sample data and is either rejected or not rejected, the alternative hypothesis is accepted only when there is sufficient statistical evidence against the null hypothesis.

3. Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

- The significance level, denoted by Œ± (alpha), is a predefined threshold in hypothesis testing that represents the maximum probability of making a Type I error, that is, rejecting a true null hypothesis. Commonly chosen values of Œ± are 0.05 or 0.01, indicating a 5% or 1% risk of incorrectly rejecting the null hypothesis. Its role in deciding the outcome of a test is to provide a decision rule: if the calculated p-value is less than or equal to Œ±, the null hypothesis is rejected; if the p-value is greater than Œ±, the null hypothesis is not rejected. Thus, the significance level controls the strictness of the test and helps ensure that conclusions drawn from sample data are statistically reliable.

4. What are Type I and Type II errors? Give examples of each.Answer ?

- Type I and Type II errors are mistakes that can occur in hypothesis testing when making decisions based on sample data.

A Type I error occurs when the null hypothesis is rejected even though it is actually true. In other words, it is a false positive. For example, suppose the null hypothesis states that a new medicine has no effect. If a researcher concludes that the medicine is effective when, in reality, it is not, this decision represents a Type I error.

A Type II error occurs when the null hypothesis is not rejected even though it is false. This is known as a false negative. For example, if the null hypothesis claims that the medicine has no effect, but the researcher fails to detect its effectiveness and concludes that it does not work, this outcome is a Type II error.

```python
# Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)

data = np.random.normal(loc=50, scale=5, size=1000)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)

margin_of_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print("Sample Mean:", round(sample_mean, 3))
print("95% Confidence Interval:", (round(ci_lower, 3), round(ci_upper, 3)))

plt.hist(data, bins=30)
plt.axvline(sample_mean)
plt.axvline(ci_lower)
plt.axvline(ci_upper)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.show()
```

```python
# Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std
    return z_scores

data = np.array([49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5])

z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=10)
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores")
plt.show()
```
