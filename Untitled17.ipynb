Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

- Anomaly Detection is the process of identifying data points or patterns that deviate significantly from expected behavior in a dataset. These anomalies often indicate important events such as fraud, faults, or rare occurrences. There are three main types of anomalies. Point anomalies are individual data points that differ greatly from the rest, such as an unusually large credit card transaction. Contextual anomalies depend on the surrounding context, meaning a value may be normal in one situation but abnormal in another, for example high electricity usage during midnight hours. Collective anomalies occur when a group of data points behaves abnormally together, even if individual points may appear normal, such as a sudden continuous spike in network traffic indicating a possible cyber attack.

Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

- Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are different approaches used for anomaly detection based on distinct principles. Isolation Forest works by randomly partitioning the data and isolating anomalies, making it efficient and suitable for high-dimensional datasets and large-scale applications. DBSCAN is a density-based clustering algorithm that identifies anomalies as points that lie in low-density regions, making it useful for spatial data and discovering clusters of arbitrary shapes. Local Outlier Factor measures how isolated a point is with respect to the density of its neighbors, making it effective for detecting local anomalies where data density varies. Each method has its strengths, with Isolation Forest being fast and scalable, DBSCAN being effective for clustering problems, and LOF being suitable for detecting subtle local deviations.

Question 3: What are the key components of a Time Series? Explain each with one example.

- A time series is composed of several key components that describe how data behaves over time. The trend component represents the long-term direction of the data, such as a steady increase in sales over years. Seasonality refers to repeating patterns that occur at regular intervals, such as higher ice cream sales during summer months. The cyclical component captures long-term fluctuations that do not follow a fixed pattern, such as economic boom and recession cycles. The residual or noise component consists of random variations that cannot be explained by the other components, such as sudden unexpected spikes or drops in stock prices. Understanding these components helps in analyzing and forecasting time series data effectively.

Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?
- A time series is considered stationary when its statistical properties, including mean, variance, and autocorrelation, remain constant over time. Stationarity is important because many forecasting models assume stable data patterns. To test whether a series is stationary, statistical tests such as the Augmented Dickey-Fuller (ADF) test or KPSS test are commonly used. If the data is found to be non-stationary, it can be transformed into a stationary series using methods like differencing to remove trends, logarithmic transformation to stabilize variance, detrending to eliminate long-term patterns, and seasonal adjustment to remove repeating seasonal effects. These transformations make the data suitable for modeling and improve forecasting accuracy.

Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.
- AR, MA, ARIMA, SARIMA, and SARIMAX are widely used time series forecasting models that differ in complexity and application. The AutoRegressive (AR) model predicts future values based on past observations of the series, while the Moving Average (MA) model uses past forecast errors for prediction. ARIMA combines AR and MA components along with differencing to handle non-stationary data. SARIMA extends ARIMA by incorporating seasonal components, making it suitable for data with repeating patterns such as monthly or quarterly trends. SARIMAX further enhances SARIMA by including exogenous variables, such as weather or external factors, which influence the time series. These models are chosen based on the nature of the data and the presence of trends, seasonality, or external influences.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Create synthetic dataset
np.random.seed(0)
time = np.arange(100)

trend = time * 0.5
seasonal = 10 * np.sin(time * 0.2)
noise = np.random.normal(0, 2, 100)

data = trend + seasonal + noise

df = pd.DataFrame({'value': data})
df.index = pd.date_range(start='2020-01-01', periods=100)

# Plot original series
df['value'].plot(title='Synthetic Time Series')
plt.show()

# Decomposition
result = seasonal_decompose(df['value'], model='additive', period=12)
result.plot()
plt.show()
```

```python
# Question 7: Apply Isolation Forest on a numerical dataset to detect anomalies and visualize them.
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

df_if = df.copy()

model = IsolationForest(contamination=0.05)
df_if['anomaly'] = model.fit_predict(df_if[['value']])

plt.scatter(df_if.index, df_if['value'], c=df_if['anomaly'])
plt.title("Isolation Forest Anomalies")
plt.show()
```

```python
# Question 8: Train a SARIMA model and forecast future values
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(df['value'], order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit()

forecast = results.forecast(steps=12)

plt.plot(df['value'], label='Original')
plt.plot(forecast, label='Forecast')
plt.legend()
plt.show()
```

```python
# Question 9: Apply Local Outlier Factor (LOF)
from sklearn.neighbors import LocalOutlierFactor
import numpy as np

X = df['value'].values.reshape(-1,1)

lof = LocalOutlierFactor(n_neighbors=20)
y_pred = lof.fit_predict(X)

plt.scatter(range(len(X)), X, c=y_pred)
plt.title("LOF Anomalies")
plt.show()
```
