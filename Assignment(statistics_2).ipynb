Question 2: What are the types of random variables?

- There are two types of random variables
- A continuous random variable is one that can take any value within a given range. These values are not countable and are usually measured rather than counted. Continuous random variables can take fractional and decimal values, making them suitable for representing physical quantities. Examples include the height of a person, the weight of an object, and the time taken to complete a task.

- A discrete random variable is one that can take only specific, countable values. These values are usually whole numbers and are obtained by counting. Since the possible outcomes can be listed individually, discrete random variables are often used in situations involving counts. Examples include the number of students present in a class, the number of heads obtained when a coin is tossed several times, or the number of defective items in a batch.

Question 3: Explain the difference between discrete and continuous distributions.

- A discrete distribution is used when the random variable can take only countable and distinct values. The probabilities are assigned to each possible value of the random variable, and the sum of all probabilities is equal to 1. Discrete distributions are generally represented using a probability mass function (PMF). Examples include the distribution of the number of students in a class, the number of heads in a fixed number of coin tosses, or the number of defective items in a batch.

- A continuous distribution is used when the random variable can take any value within a continuous range. Instead of assigning probabilities to specific values, probabilities are assigned over intervals using a probability density function (PDF). The probability of the random variable taking any exact single value is zero; only intervals have positive probability. Examples include distributions of height, weight, time, and temperature.

Each trial has only two possible outcomes, usually called success and failure, and the probability of success remains constant for every trial. - In probability, the binomial distribution is used to calculate the likelihood of getting a specific number of successes out of repeated trials. It is commonly applied in situations such as counting the number of heads in multiple coin tosses, the number of defective items in a sample, or the number of students who pass an exam. Because of its simplicity and wide applicability, the binomial distribution is one of the most important tools in probability and statistics.

Question 5: What is the standard normal distribution, and why is it important?

- The standard normal distribution is a special type of normal distribution with a mean of 0 and a standard deviation of 1. It is symmetric and bell-shaped, where most of the values are concentrated around the mean, and the probabilities decrease as we move away from the center. Any normal distribution can be converted into the standard normal distribution by using z-scores, which measure how far a value is from the mean in terms of standard deviations.

- The standard normal distribution is important because it allows us to compare values from different normal distributions using a common scale. It is widely used in probability and statistics to find probabilities, percentiles, and critical values in hypothesis testing. Many statistical methods, confidence intervals, and tests are based on the standard normal distribution, making it a fundamental concept for data analysis and decision-making.

Question 7: What is the significance of confidence intervals in statistical analysis?

- The significance of confidence intervals in statistical analysis lies in their ability to provide a range of values within which a population parameter (such as the mean or proportion) is likely to lie. Instead of giving a single estimate, a confidence interval expresses the uncertainty associated with sample data. For example, a 95% confidence interval means that if we were to take many samples and construct intervals in the same way, about 95% of those intervals would contain the true population parameter.

Question 8: What is the concept of expected value in a probability distribution?

- The expected value in a probability distribution is the long-run average outcome of a random variable if an experiment is repeated a large number of times. It represents the theoretical mean of the distribution and is calculated by multiplying each possible value of the random variable by its corresponding probability and then summing all these products. For a discrete random variable, the expected value gives a single numerical measure that summarizes the center of the distribution.

- The concept of expected value is important because it helps in decision-making under uncertainty. It is widely used in statistics, economics, gambling, insurance, and data science to predict average outcomes and compare different probabilistic scenarios. By using expected value, we can evaluate risks, returns, and overall performance when outcomes are uncertain.



```python
# Question 9: Write a Python program to generate 1000 random numbers from a normaldistribution with mean = 50 and standard deviation = 5. Compute its mean and standarddeviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=50, scale=5, size=1000)

mean = np.mean(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

plt.hist(data, bins=30)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normal Distribution (Mean=50, Std Dev=5)")
plt.show()
```

```python
# Question 10: You are working as a data analyst for a retail company.
# The company hascollected daily sales data for 2 years and wants you to identify the overall sales trend.daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,235, 260, 245, 250, 225, 270, 265, 255, 250, 260]‚óè Explain how you would apply the Central Limit Theorem to estimate the average saleswith a 95% confidence interval.
# Write the Python code to compute the mean sales and its confidence interval.

import numpy as np
import scipy.stats as stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

data = np.array(daily_sales)

n = len(data)
mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)
sem = std_sales / np.sqrt(n)

confidence_interval = stats.norm.interval(0.95, loc=mean_sales, scale=sem)

print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval:", confidence_interval)
```
