1. What is the difference between descriptive statistics and inferential statistics? Explain with examples.

- Descriptive statistics is a type of statistics used when we have to evaluate the whole population data.
- example - calculating average heights of the student of a class.

- inferential statistics comes in picture when its not possible to evalute population data due to time and lack of resources, then we use sample data and conclude result for whole population.
- example - calculating average weight of india.

2. What is sampling in statistics? Explain the differences between random
and stratified sampling.

- sampling in statistics is the process of selecting sample data from the population data in inferential statistics.

- Random sampling - here all the data points have the equal chances of getting selected for sample data but chances of creating biasness by taking more data points from the specific area.
- example - taking people average for weight in india , its possible that it take people from more crowded states.

- stratified sampling - here levels or strat are created to avoid the biasness as each level has the same probability of getting selected. it uses random sampling as sub method.

3. Define mean, median, and mode. Explain why these measures of central
tendency are important.

- Mean is the average of all values in a dataset, Median is the middle value when data is arranged in order, and Mode is the value that occurs most frequently; these measures of central tendency are important because they summarize large datasets into a single representative value, making data easier to understand, compare, and analyze.
