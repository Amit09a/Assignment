Question 5: What is the method of least squares in linear regression? one paragrpah way answer.

- The method of least squares in linear regression is a statistical technique used to estimate the regression coefficients by fitting the best possible straight line to the given data. It works by minimizing the sum of the squares of the residuals, where a residual is the difference between the observed value and the predicted value of the dependent variable. By squaring the residuals, both positive and negative errors are treated equally and larger errors are penalized more heavily. The values of the regression coefficients obtained through this method ensure that the fitted line represents the closest approximation to the actual data points, making predictions as accurate and unbiased as possible.

Question 6: What is Logistic Regression? How does it differ from Linear Regression?


- Logistic Regression is a supervised learning algorithm used for binary classification problems, where the dependent variable takes categorical values such as 0 or 1 (e.g., pass/fail, yes/no). Instead of predicting a continuous value, it predicts the probability that an observation belongs to a particular class. Logistic regression uses the sigmoid (logistic) function to map any real-valued input into a range between 0 and 1.

- In contrast, Linear Regression is used to predict continuous numerical values and assumes a linear relationship between the independent and dependent variables. Linear regression fits a straight line to the data, whereas logistic regression fits an S-shaped curve and uses a decision boundary for classification. Additionally, linear regression typically uses least squares for estimation, while logistic regression uses maximum likelihood estimation.

- Mean Absolute Error (MAE)
MAE measures the average absolute difference between the actual and predicted values. It gives an idea of how far predictions are from the true values on average and is easy to interpret because it is in the same units as the target variable.

- Mean Squared Error (MSE) / Root Mean Squared Error (RMSE)
MSE calculates the average of the squared differences between actual and predicted values, which penalizes larger errors more heavily. RMSE is the square root of MSE and is often preferred because it is in the same units as the dependent variable.

Question 8: What is the purpose of the R-squared metric in regression analysis?

- The **purpose of the R-squared ((R^2)) metric** in regression analysis is to measure how well the regression model explains the variability in the dependent variable. It represents the **proportion of the total variation in the response variable that is explained by the independent variable(s)** in the model. An (R^2) value closer to 1 indicates a better fit, meaning the model explains most of the variation in the data, while a value closer to 0 indicates a weak explanatory power. Thus, (R^2) helps in assessing the **goodness of fit** of a regression model and comparing different regression models on the same dataset.
