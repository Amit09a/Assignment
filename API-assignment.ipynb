Question 1 : What is a RESTful API?
- A RESTful API is a web service that follows the principles of REST, where the server exposes data as resources identified by URLs, and clients interact with these resources using standard HTTP methods like GET, POST, PUT, and DELETE; it is stateless, meaning each request is independent, and it typically exchanges data in JSON format, making it simple, scalable, and widely used for communication between frontend and backend systems.

Question 2: What is Flask, and why is it popular for building APIs?

- Flask is a lightweight Python web framework that makes it easy to build APIs and web applications because it is simple, flexible, and minimalistic; developers like it because it requires very little setup, supports rapid development, allows you to add only the components you need through extensions, and is easy to learn while still being powerful enough to build production-ready RESTful APIs.

Question 3:What are HTTP methods used in RESTful APIs?

- HTTP methods in RESTful APIs are standard actions used to interact with server resources, including GET (retrieve data), POST (create new data), PUT (update existing data fully), PATCH (update data partially), and DELETE (remove data), with each method performing a specific CRUD operation to ensure clear and predictable communication between client and server.

Question 4: What is the purpose of the @app.route() decorator in Flask?

- The @app.route() decorator in Flask is used to map a specific URL endpoint to a Python function, meaning it tells Flask which function should run when a client accesses that URL, allowing the creation of API routes that handle requests and return responses.

Question 5: What is the role of Flask-SQLAlchemy?

- Flask-SQLAlchemy is an extension that integrates the SQLAlchemy ORM with Flask, providing an easy way to interact with databases by allowing developers to define models as Python classes and perform CRUD operations without writing raw SQL.

Question 6: How do you create a basic Flask application?

- create a basic Flask application by installing Flask, importing it into a Python file, creating a Flask app instance, and defining at least one route using @app.route(), then running the application.

Question 7:How do you return JSON responses in Flask?

from flask import jsonify
@app.route("/data")
def data():
    return jsonify({"message": "Hello", "status": "success"})

Question 8:How do you handle POST requests in Flask?

- You handle POST requests in Flask by creating a route with methods=['POST'] and then accessing the data sent by the client using request.json or request.form.
 ex - from flask import request

@app.route("/submit", methods=["POST"])
def submit():
    data = request.json
    return {"received": data}

Question 9: How do you handle errors in Flask (e.g., 404)?

- You handle errors in Flask by using error handlers with the @app.errorhandler() decorator, which lets you define custom responses for specific HTTP error codes like 404 or 500.
- ex - @app.errorhandler(404)
def not_found(error):
    return {"error": "Resource not found"}, 404

Question 10: How do you structure a Flask app using Blueprints?

- You structure a Flask app using Blueprints by splitting your application into modular components, where each Blueprint represents a separate module with its own routes, views, and logic; you define a Blueprint in a separate file, register it in the main application file using app.register_blueprint(), and this approach keeps large applications organized and easier to maintain
