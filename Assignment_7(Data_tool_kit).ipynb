1. What is NumPy, and why is it widely used in Python?

- NumPy (Numerical Python) is a core Python library used for numerical and scientific computing.
It is widely used because it provides:

- Fast multi-dimensional arrays (ndarrays)

- Vectorized operations, which are much faster than Python loops

- Mathematical and statistical functions

- Integration with machine learning, data science, and scientific libraries

2. How does broadcasting work in NumPy?

- Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding the smaller array to match the shape of the larger one.It avoids manual looping and increases computational efficiency.

Example: [1,2,3] + 5 → [6,7,8]

3. What is a Pandas DataFrame?

- A DataFrame is a 2-dimensional, tabular data structure in Pandas with labeled rows and columns.
It is similar to an Excel sheet or SQL table, used for storing and manipulating structured data.

4. Explain the use of the groupby() method in Pandas.

 -- groupby() is used to:

- Split data into groups based on a column

- Apply operations like sum, mean, count, etc.

- Combine the results into a structured output

5. Why is Seaborn preferred for statistical visualizations?

- Seaborn is preferred for statistical visualizations because it creates visually appealing, clean, and informative plots with minimal code. It integrates smoothly with Pandas DataFrames and automatically applies statistical themes, making complex visualizations easier to build.

6. What are the differences between NumPy arrays and Python lists?

- NumPy arrays are fixed-type, optimized structures designed for numerical computation, which makes them faster and more memory-efficient than Python lists. Lists can hold mixed data types and are more flexible, but they lack built-in support for vectorized mathematical operations, making them slower for scientific tasks.

7. What is a heatmap, and when should it be used?

- A heatmap is a graphical representation of data where values are shown using colors. It is typically used to display correlations, matrix data, or patterns in large datasets, making it easy to spot relationships or clusters visually.

8. What does the term “vectorized operation” mean in NumPy?

- A vectorized operation refers to performing an arithmetic or logical operation on an entire array at once, without writing loops. NumPy uses highly optimized C code internally, allowing these operations to be extremely fast and efficient.

9. How does Matplotlib differ from Plotly?

- Matplotlib is primarily used for creating static, publication-style visualizations, while Plotly specializes in interactive graphics that allow users to zoom, hover, and explore data. Matplotlib offers more traditional, low-level control, whereas Plotly is ideal for dashboards and web-based analytics.

10. What is the significance of hierarchical indexing in Pandas?

- Hierarchical indexing, or multi-indexing, allows multiple levels of indexing on rows or columns, enabling the representation of high-dimensional data in a compact form. It simplifies tasks like slicing, aggregation, or working with grouped data structures.

11. What is the role of Seaborn’s pairplot() function?

- The pairplot() function in Seaborn creates a grid of plots that display the relationships between every pair of numerical variables in a dataset. It is mainly used for exploratory data analysis to quickly understand correlations, patterns, and distributions.

12. What is the purpose of the describe() function in Pandas?

- The describe() function provides summary statistics for numerical columns, including measures such as mean, standard deviation, minimum, maximum, and quartiles. It helps in understanding the data distribution and identifying potential issues early in analysis.

13. Why is handling missing data important in Pandas?

- Handling missing data is important because unaddressed missing values can distort analysis, introduce bias, or break machine-learning models. Pandas provides tools like dropna() to remove missing values and fillna() to replace them, ensuring clean and reliable datasets.

14. What are the benefits of using Plotly for data visualization?

- Plotly offers the advantage of creating interactive, high-quality visualizations that allow users to explore data through zooming, hovering, and toggling features. It is easy to use, supports complex charts like 3D plots and maps, and integrates well with dashboards, making it highly suitable for modern data applications.

15. How does NumPy handle multidimensional arrays?

- NumPy handles multidimensional arrays using its ndarray object, which can efficiently represent data in multiple dimensions such as 2D matrices, 3D tensors, or higher. These arrays are stored in continuous blocks of memory, allowing NumPy to perform fast computations across dimensions. With features like reshaping, slicing, broadcasting, and advanced indexing, NumPy makes working with complex multidimensional data simple and highly efficient.

16. What is the role of Bokeh in data visualization?

- Bokeh is a Python library used to create interactive and web-friendly visualizations. Unlike static plotting libraries, Bokeh allows users to zoom, hover, pan, and dynamically explore data directly in a browser. Because it integrates smoothly with HTML, JavaScript, and web frameworks like Flask or Django, it is commonly used for dashboards, live data applications, and interactive data analysis.

17. Explain the difference between apply() and map() in Pandas.

- In Pandas, map() is mainly used to apply a function element-wise to a Series, making it ideal for simple value transformations or replacements. apply(), however, is more versatile because it works on both Series and DataFrames, allowing functions to be applied to rows, columns, or entire data structures. While map() handles simple one-to-one operations, apply() is better suited for complex, custom logic involving multiple columns.

18. What are some advanced features of NumPy?

- NumPy includes several advanced features such as broadcasting, universal functions (ufuncs), structured arrays, vectorized operations, memory-mapped arrays, and high-performance linear algebra operations. It also supports Fourier transforms, random number generation, and seamless integration with optimized C and Fortran code, making it powerful for scientific computing and machine learning tasks.

19. How does Pandas simplify time series analysis?

- Pandas simplifies time series analysis by providing automatic date parsing, time-based indexing, and resampling tools that convert data into different frequencies such as daily, monthly, or yearly. It also supports rolling statistics, shifting time values, and handling irregular or missing timestamps. These built-in features allow users to perform trend analysis, forecasting preparation, and temporal filtering easily and effectively.

20. What is the role of a pivot table in Pandas?

- A pivot table in Pandas is used to summarize and transform data by turning unique values from one column into new columns while applying aggregation functions like mean, sum, or count. This helps convert long datasets into more organized and informative tables, making it easier to analyze grouped patterns and compare categories within the data.

21. Why is NumPy’s array slicing faster than Python’s list slicing?

- NumPy’s array slicing is faster because it creates a view of the original data instead of copying it. This is possible because NumPy stores arrays in contiguous memory blocks, allowing it to reference slices without additional memory usage. Python lists, however, store elements in separate memory locations, so slicing them requires creating an entirely new list, which is significantly slower.

22. What are some common use cases for Seaborn?

- Seaborn is commonly used for statistical visualizations such as distribution plots, categorical comparisons, regression analysis, correlation heatmaps, and multi-variable relationship graphs. Because it integrates well with Pandas and comes with attractive default styles, it is particularly useful during exploratory data analysis when identifying trends, patterns, and relationships in datasets.

```python
# 1. Create a 2D NumPy array and calculate the sum of each row
import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = arr.sum(axis=1)
print(row_sum)
```

```python
# 2. Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd
df = pd.DataFrame({"Marks": [88, 92, 76, 81]})
mean_val = df["Marks"].mean()
print(mean_val)
```

```python
# 4. Calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns
import pandas as pd
df = pd.DataFrame({"A": [1,2,3], "B":[3,4,5], "C":[5,6,7]})
corr = df.corr()
sns.heatmap(corr, annot=True)
```
